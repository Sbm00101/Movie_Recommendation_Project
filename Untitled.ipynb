```python
import ast
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies.genres = movies['genres'].apply(convert)
```

```python
movies.keywords = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies.cast[0]
```

```python
def convert4(obj):
    L = []
    counter = 0 
    for i in ast.literal_eval(obj):
        if (counter != 4) :
            L.append(i['name'])
            counter += 1
        else :
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert4)
```

```python
movies.head()
```

```python
movies.crew[0]
```

```python
def fetch_director(obj):
    L = []
    counter = 0 
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director' :
            L.append(i['name'])
            counter += 1
            break
    return L
```

```python
movies['director'] = movies.crew.apply(fetch_director)
```

```python
movies.drop('crew', inplace = True, axis = 1)
```

```python
movies
```

```python
movies['overview']
```

```python
movies.overview = movies['overview'].apply(lambda x:x.split())
```

```python
#Removing spaces
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['director'] = movies['director'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['director'] 
```

```python
new_df = movies.drop(columns = ['overview','genres','keywords','cast','director'], axis = 1)
```

```python
new_df
```

```python
new_df.tags = new_df.tags.apply(lambda x:" ".join(x))
```

```python
new_df.tags = new_df.tags.apply(lambda x:x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
#stem('actor acting acted he is a good guy act put on')
```

```python
new_df.tags = new_df.tags.apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
vectors = cv.fit_transform(new_df.tags).toarray()
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])), key = lambda x: x[1], reverse = True)
```

```python
def reccomend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), key = lambda x: x[1], reverse = True)[1:6]
    
    for i in movies_list :
        print(new_df.iloc[i[0]].title)
```

```python
reccomend('Avatar')
```

```python
reccomend('The Avengers')
```
